```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cardekho_dataset.csv')
df
```

```python
# Check Null values
df.isnull().sum()
```

```python
# Drop unnecessary columns
df.drop(columns=['car_name','brand','Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df.model.unique()
```

```python
# Numerical features 
numerical_features = df.select_dtypes(exclude='object').columns
print(f'Number of numerical features are : {len(numerical_features)}')

# Categorical features
categorical_features = df.select_dtypes(include='object').columns
print(f'Number of categorical features are : {len(categorical_features)}')

# Discrete features 
discrete_features = [feature for feature in numerical_features if len(df[feature].unique())<=25 ]
print(f'NUmber of discrete features : {len(discrete_features)}')

# Continuous features
continuous_features = [feature for feature in numerical_features if feature not in discrete_features]
print(f'Number of continuous features : {len(continuous_features)}')
```

```python
# Seperating the dependent and independent features 
X = df.drop('selling_price', axis=1)
y = df['selling_price']
```

```python
df['model'].value_counts()
```

```python
len(df['model'].unique())
```

```python
# Feature Encoding for Categorical data
# Label encoding for model 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['model'] = le.fit_transform(X["model"])
```

```python
# Onehot encoding for other categorical features
# Scaling the numerical features using standard scaler

num_features = X.select_dtypes(exclude='object').columns
oh_columns = ['seller_type','fuel_type','transmission_type']

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('OneHot Encoder',OneHotEncoder(drop='first'),oh_columns),
        ('StandardScaler',StandardScaler(), num_features),
    ], remainder='passthrough' # Keep all columns that are not explicitly transformed and pass them through to the output unchanged.
)
```

```python
X = preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(
    data = np.hstack((X, y.values.reshape(-1, 1))),
     columns = [f'Feature{i+1}' for i in range(X.shape[1])] + ['Target']
).to_csv('data.csv', index=False)
```
