```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('weight_height.csv')
df.head()
```

```python
# Converting meter into centimeter
df['Height (cm)'] = df['Height (m)'].apply(lambda x: x * 100)
```

```python
plt.scatter(df['Weight (kg)'],df['Heigh (cm)'])
plt.xlabel('Weight (kg)')
plt.ylabel('Heigh (cm)')
```

```python
# Correlation between them is quite moderate
correlation = df['Height (cm)'].corr(df['Weight (kg)'])
print(f"Correlation between Height and Weight: {correlation}")
```

```python
# Plot heatmap of the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
```
