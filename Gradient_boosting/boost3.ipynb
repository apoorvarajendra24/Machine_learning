```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import log_loss, accuracy_score
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
import numpy as np 
```

```python
sonar = pd.read_csv("Sonar.csv")
sonar
```

```python
sonar.info()
```

```python
sonar.describe()
```

```python
sonar.isnull().sum()
```

```python
y = sonar['Class']
X = sonar.drop('Class', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=24, stratify=y)
```

```python
model = GradientBoostingClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(f'Predicted values : {y_pred}')
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
params = {'learning_rate': np.linspace(0.001, 0.9, 10),
          'max_depth': [None, 2, 3, 4],
          'n_estimators': [25, 50, 100]}
kfold = StratifiedKFold(n_splits=5, shuffle=True,
                        random_state=24)
gbm = GradientBoostingClassifier(random_state=24)
gcv = GridSearchCV(gbm, param_grid=params, cv=kfold,
                   scoring='neg_log_loss')
gcv.fit(X, y)
pd_cv = pd.DataFrame( gcv.cv_results_ )
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
y_pred_gcv = gcv.predict(X_test)
print(accuracy_score(y_test,y_pred_gcv))
```
